```python
import pandas as pd
import numpy as np
```

```python
mydict = {}
df = pd.DataFrame()
with open('amazon-meta.txt') as f:
    for line in f:
        if line.startswith('Id:'):
            df = df.append(mydict, ignore_index = True)
            mydict = {}
            info = line.split(':', 1)
            mydict[info[0]] = info[1].rstrip('\n')
        elif (line.startswith('ASIN:')):
            info = line.split(':', 1)
            mydict[info[0]] = info[1].rstrip('\n')
        elif (line.startswith('  title:') |
              line.startswith('  group:') |
              line.startswith('  salesrank:')):
            info = line.split(':', 1)
            mydict[info[0][2:]] = info[1].rstrip('\n')
        elif line.startswith('  similar:'):
            info = line.split(':', 1)
            mydict[info[0][2:]] = info[1][4:].rstrip('\n').split('  ')
        elif line.startswith('  categories:'):
            info = line.split(':', 1)
            mydict[info[0][2:]] = [ ]
        elif line.startswith('   |'):
            mydict[info[0][2:]].append(line[3:].rstrip('\n'))
        elif line.startswith('  reviews:'):
            info = line.split(':', 1)
            mydict[info[0][2:]] = [ ]
        elif line.startswith('    '):
            mydict[info[0][2:]].append(line[4:].rstrip('\n'))

df.to_csv('am.csv')
```

```python
def reviewdict(review):
    if type(review) == float:
        return None
    toreturn = []
    for line in review:
        thisdict = {}
        a = line.split(' ')
        a = [x for x in a if x!='']
        thisdict['date'] = a[0]
        thisdict['customer'] = a[2]
        thisdict['rating'] = a[4]
        thisdict['votes'] = a[6]
        thisdict['helpful'] = a[8]
        toreturn.append(thisdict)
    return toreturn
```

```python
reviewlistofdicts = []
for review in df.reviews:
    parsedreview = reviewdict(review)
    if parsedreview != None:
        reviewlistofdicts += (parsedreview)

reviewdf = pd.DataFrame(reviewlistofdicts)
```

```python
def catsplit(categories):
    if type(categories) == float:
        return None
    toreturn = []
    for line in categories:
        thisdict = {}
        a = line.split('|')
        a = [x for x in a if x!='']
        toreturn.append(a)
    return toreturn
```

```python
cats = []
for line in df.categories:
    if type(line) != float:
        cats += catsplit(line)
```
