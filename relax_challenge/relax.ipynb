```python
import pandas as pd
%matplotlib inline
```

```python
users = pd.read_csv('takehome_users.csv')
```

```python
users
```

```python
engagement = pd.read_csv('takehome_user_engagement.csv')
```

```python
users = users.set_index('object_id')

users = users.join(engagement.groupby('user_id').sum() >2)
```

```python
engagement.time_stamp = pd.DatetimeIndex(engagement.time_stamp)
```

```python
gb = engagement.groupby(['user_id', 'time_stamp']).count()
```

```python
users.visited = users.visited.fillna(False)
```

```python
threex = users[users.visited]
```

```python
threex['adopted'] = False
threex
```

```python
for user in threex.index:
    for idx in range(len(gb.ix[user].index[:-2])):
        if (gb.ix[user].index[idx+2] - gb.ix[user].index[idx]) <= pd.Timedelta('7 days'):
            threex.loc[user, 'adopted'] = True
            break
```

```python
threex
```

```python
users = users.join(threex.adopted)
users
```

```python
relevant = pd.get_dummies(users.creation_source).join(users[['opted_in_to_mailing_list', 'enabled_for_marketing_drip', 'adopted']])
relevant.adopted = relevant.adopted.fillna(False)
relevant
```

```python
import seaborn as sns
relevant.corr()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(relevant.drop('adopted', axis = 1), relevant.adopted)
```

```python
clf.coef_
```

The features that seem to have some predictive power for whether a user is adopted are the creation source (specifically, users who signed up via a guest invite or google_auth are more likely to stick around) and whether the user is signed up for the mailing list or marketing drip.
