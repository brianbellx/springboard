```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_json('logins.json')
```

```python
data['tally'] = 1
dfg = data.groupby(pd.Grouper(key='login_time', freq='15T')).sum()
```

```python
dfg = dfg.fillna(0)
rolldfg = dfg.rolling(3, center = True ).mean()
rolldfg.loc['1970-01-05']
```

```python
fig = plt.figure(figsize=(20,10))
fig = plt.plot(rolldfg.ix['1970-02-02':'1970-02-08'])
```
